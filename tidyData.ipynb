```python
import re
def clean_phone(phone_number):
    phone_number = re.sub('[^\w\s]', '', phone_number)
    phone_number = re.sub(' |x', '', phone_number)
    return phone_number
```

```python
revenue_custinfo['phone2'] = revenue_custinfo['phone'].apply(clean_phone)
```

```python
revenue_custinfo['phone_len'] = revenue_custinfo['phone2'].apply(len)
```

```python
def is_prime(n):
    primes = []
    if n == 1:
        primes.append(1)
    elif n == 2:
        primes.append(1, 2)
    else:
        prime = True
        for i in range(2, n):
            if n % i == 0:
                prime = False
                break
    return prime
```

```python
is_prime(97)
```

```python
import math
def find_primes(n):
    primes = []
    if n == 1:
        primes.append(1)
    else:
        for i in range(2, n):
            prime = True
            #for j in range(2, math.floor(math.sqrt(i))):
            for j in range(2, i):
                if i % j == 0:
                    prime = False
                    break
            if prime == True:
                primes.append(i)
    return primes
```

```python
import timeit
start = timeit.default_timer()
primes = find_primes(9999)
end = timeit.default_timer()
print(end-start)
print(primes)
```
